```python
import pandas as pd
import matplotlib
import numpy as np
%matplotlib inline

tab_list = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8')

tab_list[12]
```

```python
tab_list[12].head()
```

```python
dfs=tab_list[12]
rows, columns = dfs.shape
print(f"Кількість рядків: {rows}")
print(f"Кількість стовпців: {columns}")
```

```python
# Замінюємо значення "—" на NaN у датафреймі
dfs.replace('—', float('NaN'), inplace=True)
dfs

```

```python
# типи всіх стовпців
dfs.dtypes
```

```python
# Замінюємо типи нечислових колонок на числові
dfs = dfs.apply(pd.to_numeric, errors='ignore')
dfs.dtypes
```

```python
# Обчислити частку пропусків у кожній колонці
null_counts = dfs.isnull().sum()
total_rows = dfs.shape[0]
null_fractions = null_counts / total_rows
null_fractions
```

```python
# Видаляємо останній рядок таблиці
dfs = dfs.drop(dfs.index[-1])
dfs
```

```python
# Замінюємо пропуски середніми значеннями лише для числових стовпців
numeric_columns = dfs.select_dtypes(include=['float64', 'int64']).columns
dfs[numeric_columns] = dfs[numeric_columns].fillna(dfs[numeric_columns].mean())
dfs
```

```python
# Обчислюємо середнє значення народжуваності по Україні у 2019 році
average_2019 = dfs['2019'].mean()

# Фільтруємо регіони, де рівень народжуваності у 2019 році був вищим за середній по Україні
regions_higher_than_average_2019 = dfs[dfs['2019'] > average_2019]

# Отримуємо список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
regions_list = regions_higher_than_average_2019['Регіон'].tolist()

print(regions_list)
```

```python
# Знаходимо регіон з найвищою народжуваністю у 2014 році
highest_birth_rate_2014 = dfs.loc[dfs['2014'].idxmax(), 'Регіон']

print("Регіон з найвищою народжуваністю у 2014 році:", highest_birth_rate_2014)
```

```python
import matplotlib.pyplot as plt

# Вибираємо дані для побудови діаграми
birth_rates_2019 = dfs['2019']
regions = dfs['Регіон']

# Побудова стовпчикової діаграми
plt.figure(figsize=(10, 6))
plt.bar(regions, birth_rates_2019, color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Кількість народжень у 2019 році')
plt.title('Народжуваність по Україні у 2019 році')
plt.xticks(rotation=90)  # Повертаємо назви регіонів для кращої видимості
plt.tight_layout()
plt.show()
```

```python
birth_rates_2019 = dfs['2019']
regions = dfs['Регіон']

# Побудова кругової діаграми
plt.figure(figsize=(8, 8))
plt.pie(birth_rates_2019, labels=regions, autopct='%1.1f%%', startangle=140)
plt.title('Народжуваність по Україні у 2019 році')
plt.axis('equal')  # Забезпечує круглу форму діаграми
plt.show()
```

```python
birth_rates_2019 = dfs['2019']
regions = dfs['Регіон']

# Побудова кольорового графіку
plt.figure(figsize=(10, 8))
plt.scatter(regions, birth_rates_2019, c=birth_rates_2019, cmap='coolwarm', marker='s', s=100)
plt.xticks(rotation=90)  # Повертаємо назви регіонів для кращої читабельності
plt.xlabel('Регіон')
plt.ylabel('Народжуваність')
plt.title('Народжуваність по регіонах України у 2019 році')
plt.colorbar(label='Народжуваність')
plt.grid(True, linestyle='--', alpha=0.7)  # Додаємо сітку
plt.tight_layout()
plt.show()
```
