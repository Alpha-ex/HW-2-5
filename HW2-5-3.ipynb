```python
# Максимальна ціна
max_price = dfs['price'].max()

# Мінімальна ціна
min_price = dfs['price'].min()

# Середня ціна
mean_price = dfs['price'].mean()

# Медіанна ціна
median_price = dfs['price'].median()

# Виведення результатів
print('Відповідь:')
print("Максимальна ціна:", max_price)
print("Мінімальна ціна:", min_price)
print("Середня ціна:", mean_price)
print("Медіанна ціна:", median_price)
```

```python
# Знаходження найвищого рейтингу
max_rating = dfs['user_rating'].max()

# Виведення результату
print("Відповідь - Найвищий рейтинг = ", max_rating)
```

```python
# Знаходимо кількість книг з найвищим рейтингом
num_books_with_rating = dfs[dfs['user_rating'] == 4.9].shape[0]

# Виведення результату
print("Кількість книг з найвищим рейтингом:", num_books_with_rating)
```

```python
# Відфільтруємо дані для року 2015
books_2015 = dfs[dfs['year'] == 2015]

# Знаходимо найдорожчу книгу серед книг за 2015 рік
most_expensive_book_2015 = books_2015.loc[books_2015['price'].idxmax()]

# Виводимо результат
print('Відповідь: З тих книг, що потрапили до Топ-50 у 2015 році, найдорожча книга - ')
most_expensive_book_2015
```

```python
# Відфільтруємо дані для книг жанру Fiction та року 2010
fiction_books_2010 = dfs[(dfs['genre'] == 'Fiction') & (dfs['year'] == 2010)]

# Знаходимо кількість книг жанру Fiction, які потрапили до Топ-50 у 2010 році
num_fiction_books_2010 = len(fiction_books_2010)

# Виводимо результат
print("Відповідь - кількість книг жанру Fiction, які потрапили до Топ-50 у 2010 році:", num_fiction_books_2010)
```

```python
# Відфільтруємо дані для книг з рейтингом 4.9 та роками 2010 і 2011
books = dfs[(dfs['user_rating'] == 4.9) & (dfs['year'].isin([2010, 2011]))]

# Знаходимо кількість таких книг
num_books = len(books)

# Виводимо результат
print("Відповідь - кількість книг з рейтингом 4.9, які потрапили до рейтингу у 2010 та 2011 роках:")
num_books
```

```python

# Згрупуємо дані за жанр та застосуємо функції агрегації
genre_price_stats = dfs.groupby('genre').agg({'price': ['min', 'max']})

# Перейменуємо стовпці
genre_price_stats.columns = ['min_price', 'max_price']

# Виведемо результат
genre_price_stats
```

```python
# Згрупуємо дані за автором та жанром
author_genre_count = dfs.groupby(['author', 'genre']).agg({'name': 'count'})

# Перейменуємо стовпець 'name' на 'book_count'
author_genre_count = author_genre_count.rename(columns={'name': 'book_count'})

# Скинути індекси і видалити їх з виводу
author_genre_count = author_genre_count.reset_index()

# Виберемо книги з необхідними жанрами
author_genre_count = author_genre_count[author_genre_count['genre'].isin(['Fiction', 'Non Fiction'])]

author_genre_count
```

```python
author_genre_count.shape
```

```python
# Знайдемо автора з найбільшою кількістю книг
author_most_books = author_genre_count.loc[author_genre_count['book_count'].idxmax()]
print("Відповідь - автор з найбільшою кількістю книг:")
author_most_books
```

```python
# Згрупуємо дані за автором та жанром
author_genre_mean_rating = dfs.groupby(['author', 'genre']).agg({'user_rating': 'mean'})

# Перейменуємо стовпець 'user_rating' на 'mean_rating'
author_genre_mean_rating = author_genre_mean_rating.rename(columns={'user_rating': 'mean_rating'})

# Скинути індекси
author_genre_mean_rating = author_genre_mean_rating.reset_index()

# Виведемо результат
author_genre_mean_rating
```

```python
# Знайдемо індекс рядка з мінімальним середнім рейтингом
min_mean_rating_index = author_genre_mean_rating['mean_rating'].idxmin()

# Отримаємо ім'я автора з найменшим середнім рейтингом
author_min_mean_rating = author_genre_mean_rating.iloc[min_mean_rating_index]

# Отримаємо середній рейтинг цього автора
min_mean_rating = author_genre_mean_rating.loc[min_mean_rating_index, 'mean_rating']

# Виведемо результат
print("Відповідь - автор з найменшим середнім рейтингом - ", author_min_mean_rating)
min_mean_rating
```

```python
# З'єднаємо два датафрейми по стовпцях
merged_df = pd.concat([author_genre_count, author_genre_mean_rating], axis=1)

# Виведемо результат
merged_df
```

```python
# Сортуємо датафрейм за кількістю книг та рейтингом за зростанням
sorted_df = merged_df.sort_values(by=['book_count', 'mean_rating'], ascending=True)

sorted_df
```

```python
# Отримання імені першого автора у відсортованому списку
first_author = sorted_df.loc[1]

# Виведення результату
print("Відповідь - перший автор у списку:")
first_author
```

```python
# Групуємо дані по року видання та обчислюємо середній рейтинг
yearly_mean_rating = dfs.groupby('year')['user_rating'].mean()

# Побудова стовпчастої діаграми
yearly_mean_rating.plot(kind='bar', color='skyblue', edgecolor='black')

# Налаштування заголовка та підписів осей
plt.title('Залежність середнього рейтингу від року видання книги')
plt.xlabel('Рік видання')
plt.ylabel('Середній рейтинг')
plt.xticks(rotation=45)  # Обертання підписів осі X на 45 градусів

# Відображення графіка
plt.tight_layout()
plt.show()
```

```python
# Підрахунок кількості книг у кожному жанрі
genre_counts = dfs['genre'].value_counts()

# Побудова кругової діаграми
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightblue'])

# Налаштування заголовка
plt.title('Розподіл жанрів книг')

# Відображення графіка
plt.show()
```

```python
# Групування даних за рік та обчислення середнього рейтингу
yearly_mean_rating = dfs.groupby('year')['user_rating'].mean()

# Побудова лінійного графіка
plt.figure(figsize=(10, 6))
plt.plot(yearly_mean_rating.index, yearly_mean_rating.values, marker='o', linestyle='-')

# Налаштування заголовка та міток осей
plt.title('Середній рейтинг книг за роками')
plt.xlabel('Рік')
plt.ylabel('Середній рейтинг')

# Відображення графіка
plt.grid(True)
plt.show()
```

```python
# Побудова ящика з вусами
plt.figure(figsize=(10, 6))
sns.boxplot(x=dfs['user_rating'], color='skyblue')

# Налаштування заголовка та міток осей
plt.title('Розподіл рейтингів книг')
plt.xlabel('Рейтинг')
plt.ylabel('Кількість')

# Відображення графіка
plt.grid(True)
plt.show()
```

```python
# Встановлення стилю та розміру графіку
sns.set_style("darkgrid")
plt.figure(figsize=(12, 8))

# Побудова кольорового графіка
sns.scatterplot(data=dfs, x='user_rating', y='price', hue='genre', size='reviews', sizes=(20, 200))

# Налаштування заголовка та міток осей
plt.title('Зв’язок між рейтингом, ціною та жанром книг')
plt.xlabel('Рейтинг')
plt.ylabel('Ціна')

# Додавання легенди
plt.legend(title='Жанр')

# Відображення графіка
plt.show()
```
