```python
import pandas as pd

# Зчитування файлу CSV
df = pd.read_csv("2017_jun_final.csv")

# Виведення перших кількох рядків для перевірки
print(df.head())
```

```python
# Визначення розміру таблиці за допомогою методу shape
table_shape = df.shape

# Виведення розміру таблиці
print("Розмір таблиці:", table_shape)
```

```python
# Визначення типів всіх стовпців за допомогою атрибута dtypes
column_types = df.dtypes

# Виведення типів стовпців
print(column_types)
```

```python
# Порахунок частки пропусків у кожній колонці
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Виведення частки пропусків у кожній колонці
print("Частка пропусків у кожній колонці:")
print(missing_percentage)
```

```python
# Виберемо стовбець "Мова програмування"
columns_to_keep = ['Мова програмування']

# Використовуємо список для вибору потрібних стовпців
df_cleaned = df[columns_to_keep]

# Виведення перших кількох рядків очищеного DataFrame
print(df_cleaned.head())
```

```python
# Порахунок частки пропусків у кожній колонці
missing_percentage_after_removal = (df_cleaned.isnull().sum() / len(df_cleaned)) * 100

# Виведення частки пропусків у кожній колонці після видалення стовпців
print("Частка пропусків у кожній колонці після видалення стовпців:")
print(missing_percentage_after_removal)
```

```python
# Видалення усіх рядків з пропущеними значеннями
df_cleaned = df.dropna()

# Виведення перших кількох рядків очищеного DataFrame
print(df_cleaned.head())
```

```python
# Визначення нового розміру таблиці
new_table_shape = df_cleaned.shape

# Виведення нового розміру таблиці
print("Новий розмір таблиці:", new_table_shape)
```

```python
# Створення нової таблиці, де значення в стовпці "Мова програмування" рівне 'Python'
python_data = df[df['Мова програмування'] == 'Python']

# Виведення перших кількох рядків нової таблиці
print(python_data.head())
```

```python
# Визначення розміру таблиці python_data 
python_data_shape = python_data.shape

# Виведення розміру таблиці python_data
print("Розмір таблиці python_data:", python_data_shape)
```

```python
# Групування за стовпчиком "Посада"
grouped_by_position = df.groupby('Посада')

# Виведення перших кількох рядків з кожної групи
for position, group in grouped_by_position:
    print("Посада:", position)
    print(group.head())
    print("\n")
```

```python
# Агрегація даних за допомогою методу agg
aggregated_data = df.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})

# Виведення агрегованих даних
print(aggregated_data)
```

```python
def fill_avg_salary(row):
    # Повертає значення стовпця "Зарплата на місяць"
    return row.mean()

# Додавання нового стовпця "avg" за допомогою методу apply
df['avg'] = df.apply(fill_avg_salary, axis=1)

# Виведення перших кількох рядків DataFrame з новим стовпцем "avg"
print(df.head())
```

```python
# Отримання описової статистики для стовпчика "avg"
avg_stats = df['avg'].describe()

# Виведення описової статистики
print(avg_stats)
```

```python
# Збереження DataFrame у CSV файл з кодуванням UTF-8
df.to_csv('output_data_utf8.csv', index=False, encoding='utf-8')

print("Таблицю успішно збережено у файлі output_data_utf8.csv з кодуванням UTF-8")
```
