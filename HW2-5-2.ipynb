```python
import pandas as pd

# Зчитування файлу CSV з вказанням параметра sep
dfs = pd.read_csv("2017_jun_final.csv")

# Виведення перших кількох рядків для перевірки
dfs.head()
```

```python
# Виведення розміру таблиці
print("Розмір таблиці:")
dfs.shape
```

```python
# Виведення типів стовпців
dfs.dtypes
```

```python
# Порахунок пропущених значень у кожному стовпці
missing_values = dfs.isnull().sum()

# Визначення загальної кількості рядків у таблиці
total_rows = len(dfs)

# Порахунок частки пропусків у кожному стовпці
missing_ratio = missing_values / total_rows

# Виведення частки пропусків у кожному стовпці
print("Частка пропусків у кожному стовпці:")
missing_ratio
```

```python
# Виберемо стовбець "Мова програмування"
columns_to_keep = ['Мова програмування']

# Використовуємо список для вибору потрібних стовпців
dfs_cleaned = dfs[columns_to_keep]

# Виведення перших кількох рядків очищеного DataFrame
dfs_cleaned.head()
```

```python
# Виведення першого рядка очищеного DataFrame, щоб перевірити наявність лише стовпця "Мова програмування"
dfs_cleaned.head(1)

# Порахунок пропущених значень у кожному стовпці
missing_values_cleaned = dfs_cleaned.isnull().sum()

# Визначення загальної кількості рядків у таблиці
total_rows_cleaned = len(dfs_cleaned)

# Порахунок частки пропусків у кожному стовпці
missing_ratio_cleaned = missing_values_cleaned / total_rows_cleaned

# Виведення частки пропусків у кожному стовпці
print("Частка пропусків у кожному стовпці після видалення:")
missing_ratio_cleaned
```

```python
# Фільтрація рядків з мовою програмування Python
python_rows = dfs[dfs['Мова програмування'] == 'Python']

# Видалення рядків з пропусками, за винятком рядків з мовою програмування Python
dfs_cleaned = pd.concat([dfs.dropna(), python_rows])

# Виведення перших кількох рядків очищеного DataFrame
dfs_cleaned.head()
```

```python
# Визначення розміру нової таблиці
cleaned_shape = dfs_cleaned.shape

# Виведення розміру нової таблиці
print("Розмір нової таблиці:")
cleaned_shape
```

```python
# Фільтрація рядків, де мова програмування - Python
python_data = dfs[dfs['Мова програмування'] == 'Python']

# Виведення перших кількох рядків нової таблиці
python_data.head()
```

```python
python_data_shape = python_data.shape

# Виведення розміру таблиці python_data
print("Розмір таблиці python_data:")
python_data_shape
```

```python
# Групування за стовпчиком "Посада"
grouped_data = dfs.groupby('Посада')

# Виведення перших кількох рядків з кожної групи
for position, group in grouped_data:
    print("Посада:", position)
group.head()
```

```python
# Виконання агрегації даних за стовпчиком "Посада" і знаходження мінімального та максимального значення у стовпчику "Зарплата на місяць"
aggregated_data = dfs.groupby('Посада')['Зарплата на місяць'].agg(['min', 'max'])

# Виведення агрегованих даних
aggregated_data
```

```python
def fill_avg_salary(row):
    # Перетворення значень рядка в числа
    row_numeric = pd.to_numeric(row, errors='coerce')
    avg_salary = row_numeric.mean(skipna=True)
    return avg_salary


# Створення нового стовпчика "avg" за допомогою методу apply та функції fill_avg_salary
dfs['avg'] = dfs.apply(fill_avg_salary, axis=1)

# Виведення перших кількох рядків DataFrame для перевірки
dfs.head()
```

```python
# Описова статистика для стовпчика "avg"
avg_salary_stats = dfs['avg'].describe()

# Виведення описової статистики
avg_salary_stats
```

```python
# Збереження таблиці у CSV файл
dfs.to_csv('output_data.csv', index=False)

print("Таблицю успішно збережено в файлі output_data.csv")
```
